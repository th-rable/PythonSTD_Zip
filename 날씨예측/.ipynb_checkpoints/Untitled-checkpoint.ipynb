```python
def split(orgdata,n):
    x_split = []
    y_split = []
    for i in range(len(orgdata)):
        x=orgdata.loc[i:i+n-1,:]
        if (i+n)<len(orgdata):
            x_split.append(x)
            y_split.append(orgdata.loc[i+n,:])
        else:
            break
    return np.array(x_split),np.array(y_split)
    
```

```python
x,y = split(processdata,10)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,shuffle=False)
```

```python
model = keras.Sequential()
model.add(keras.layers.Dense(64,input_shape=(10,12),activation='elu'))

model.add(keras.layers.LSTM(50,activation='tanh',return_sequences=True))
model.add(keras.layers.Dropout(0.2))
model.add(keras.layers.LSTM(100,activation='tanh',return_sequences=True))
model.add(keras.layers.Dropout(0.2))
model.add(keras.layers.LSTM(200,activation='tanh',return_sequences=True))
model.add(keras.layers.Dropout(0.3))
model.add(keras.layers.LSTM(400,activation='tanh',return_sequences=False))
model.add(keras.layers.Dropout(0.3))
model.add(keras.layers.Dense(12,activation='elu'))
```

```python
model = keras.Sequential()
model.add(keras.layers.Dense(128,input_shape=(10,12),activation='elu'))
#model.add(keras.layers.Dropout(0.2))
model.add(keras.layers.LSTM(512,activation='tanh',return_sequences=False))
#model.add(keras.layers.Dropout(0.2))
model.add(keras.layers.Dense(12,activation='elu'))
```

```python

lr_schedule = keras.optimizers.schedules.ExponentialDecay(initial_learning_rate=0.001, decay_steps=10000,decay_rate=0.9)
model.compile(optimizer=keras.optimizers.Adam(learning_rate=lr_schedule), loss="mse",metrics=['accuracy'])

```

```python

model.compile(optimizer=keras.optimizers.Adam(learning_rate=0.001), loss="mse",metrics=['accuracy'])

```

```python
model.summary()
```

```python
model.fit(x_train,y_train,epochs=10,verbose=1)
```

```python
model.save('model1.h5')
```

```python
(score,loss) = model.evaluate(x_train,y_train,verbose=1)
```
